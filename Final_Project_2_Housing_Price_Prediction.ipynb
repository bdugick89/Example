##Traiage for Nulls and Zeros

```python
row_count = data_frame_extract.shape[0]
columns_with_nulls = data_frame_extract.columns[data_frame_extract.isnull().any()]
null_count = data_frame_extract[columns_with_nulls].isnull().sum()
null_percent = (data_frame_extract[columns_with_nulls].isnull().sum()/row_count) * 100
null_counts_df = pd.DataFrame({'Column': null_percent.index, 'Null Count': null_count.values, 'Null Percent': null_percent.values})
null_counts_df.sort_values(by='Null Percent', ascending=False)
```

```python
row_count = data_frame_extract.shape[0]
columns_with_zeroes = data_frame_extract.columns[(data_frame_extract == 0).any()]
zero_count = (data_frame_extract[columns_with_zeroes] == 0).sum()
zero_percent = (zero_count/row_count) * 100
zero_counts_df = pd.DataFrame({'Column': zero_count.index, 'Zero Count': zero_count.values, 'Zero Percent': zero_percent.values})
zero_counts_df.sort_values(by='Zero Percent', ascending=False)
```

### RMSE & RMSPE

### RSME Definition

### Root Mean Squared Error (RMSE):
- RMSE measures the average deviation of the predicted values from the actual values. It is calculated by taking the square root of the mean of the squared differences between the predicted values and the actual values.
- **Formula:**  
  \[ RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} \]  
  Where:
  - \( n \) is the number of samples
  - \( y_i \) is the actual value of the target for sample \( i \)
  - \( \hat{y}_i \) is the predicted value of the target for sample \( i \)

### Root Mean Squared Percentage Error (RMSPE):
- RMSPE is a variant of RMSE that is used when the target variable has a wide range of values. It calculates the percentage error between the predicted and actual values before taking the square root of the mean of the squared percentage errors.
- **Formula:**  
  \[ RMSPE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left(\frac{y_i - \hat{y}_i}{y_i}\right)^2} \]  
  Where:
  - \( n \) is the number of samples
  - \( y_i \) is the actual value of the target for sample \( i \)
  - \( \hat{y}_i \) is the predicted value of the target for sample \( i \)

This block of code can be described as hyperparameter tuning for a Ridge regression model using cross-validation to find the optimal value of the regularization parameter (alpha). It iterates over a predefined range of alpha values, trains a Ridge regression model for each alpha value, and evaluates the model's performance using cross-validation to calculate the root mean squared error (RMSE). Finally, it creates a DataFrame results_df to store the RMSE values for different alpha values and sorts it to identify the alpha value that results in the lowest RMSE.







```python
# Specify values of alpha we want to try
alph = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1e-1, 1, 2, 3, 4, 5, 10, 20, 30, 50]

# Create empty list to hold results
results_alpha = []

for i in alph:
  # Train the model
  my_fit_rr = Ridge(alpha = i) # Looping through alpha
  my_fit_rr.fit(X_train, y_train)

  # Calculate CV RMSE
  results = cross_val_score(
      my_fit_rr,
      X,
      y,
      scoring='neg_root_mean_squared_error',
      cv = 10
      )
  rmse = abs(results.mean())
  results_alpha.append(rmse)

results_df = pd.DataFrame({'Alpha': alph, 'RMSE': results_alpha})
print(results_df.sort_values(by=['RMSE']))
```

### Run With Alfa 30

```python
# Retrain the model w/ alpha = 30
my_fit_rr = Ridge(alpha = 29)
my_fit_rr.fit(X_train, y_train)

# Calculate CV RMSE
results = cross_val_score(my_fit_rr, X, y, scoring='neg_root_mean_squared_error', cv = 10)
rmse = abs(results.mean())
print(f"CV RMSE: {rmse}")
```
